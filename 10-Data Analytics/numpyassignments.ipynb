### Assignment 4: Statistical Operations

1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.

```python
arr = np.random.randint(1,11, size=(5,5))
print("Original Array is: ")
print(arr)

print("The mean of Array is: ", np.mean(arr))
print("The median of Array is: ", np.median(arr))
print("The standard deviation of Array is: ", np.std(arr))
print("The variance of the Array is: ", np.var(arr))
```

2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

```python
arr = np.arange(1,10).reshape(3,3)
print("Original Array is: ")
print(arr)

mean_arr = np.mean(arr)
std_arr = np.std(arr)

#The normalization of array
normalized_arr = (arr - mean_arr) / std_arr
print("The normalized array is: \n", normalized_arr)
```

### Assignment 9: Structured Arrays

1. Create a structured array with fields 'name' (string), 'age' (integer), and 'weight' (float). Add some data and sort the array by age.

```python
data_type = [('name', 'U20'), ('age', 'i4'), ('weight', 'f4')] #this is the custom datatype we have created where U20: string upto 20 chars, i4: int-4byte and f4: float-4byte
data = np.array([('subrat', 44, 74.4), ('raj', 25, 45.9), ('sohan', 24, 89.0)], dtype=data_type)
print(data)

sorted_data = np.sort(data, order='age')
print(sorted_data)
```

2. Create a structured array with fields 'x' and 'y' (both integers). Add some data and compute the Euclidean distance between each pair of points.

```python
data_type = [('x', 'i4'), ('y', 'i4')]
data = np.array([(1,2), (3,4), (7,8)], dtype=data_type)

print(data['x'])
print(data['y'])

#Important think to understand
distances = np.sqrt((data['x'][:, np.newaxis] - data['x'])**2 + (data['y'][:, np.newaxis] - data['y'])**2)
print(distances)
```

### Assignment 10: Masked Arrays

1. Create a masked array of shape (4, 4) with random integers and mask the elements greater than 10. Compute the sum of the unmasked elements.

```python
''' 
Masked Arrays: Masked arrays are arrays that may have missing or invalid entries. The masked array module provides a convenient way to handle such data. 
You can create a masked array by using the numpy.ma.masked_array function, which takes an array and a mask as input. The mask is a boolean array that indicates which entries are valid 
(False) and which are invalid (True).
'''

import numpy.ma as ma

arr = np.random.randint(1,21, size=(4,4))
print("Original Array: \n", arr)

masked_arr = ma.masked_array(arr, mask=arr > 10) # it will create a mask array and mask=arr>10 means those values are greater than 10 are masked means hidden/invisible
print("Masked Array: \n", masked_arr)

masked_sum = masked_arr.sum()
print("Sum of all unmasked elements: ", masked_sum)
```

2. Create a masked array of shape (3, 3) with random integers and mask the diagonal elements. Replace the masked elements with the mean of the unmasked elements.

```python
arr = np.random.randint(1,21, size=(3,3))
print("Original Array: \n", arr)

masked_arr = ma.masked_array(arr, mask=np.eye(3))
print("Masked Array: \n", masked_arr)

unmasked_mean = masked_arr.mean()
print("The mean of unmasked array: ", unmasked_mean)

masked_arr = masked_arr.filled(unmasked_mean) #If basically fill the unmasked values with the mean of masked values
print("The new masked arr: \n", masked_arr)
```
