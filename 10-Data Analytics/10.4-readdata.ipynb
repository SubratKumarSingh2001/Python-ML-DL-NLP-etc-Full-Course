## Reading Data From Different Sources 

```python
#Reading from the json 
import pandas as pd
from io import StringIO

Data = '{"employee_name": "Subrat", "email": "sxcf133@gmail.com", "job_profile": [{"title1": "Data Scientist", "title2": "Web Developer", "title3": "Data Engineer"}]}'
string_obj = StringIO(Data) #here the Data is the normal string and .read_json accepts a file or a url to the file so StringIO convert the string into the file like object, so as .read_json consider Data as a file
df = pd.read_json(string_obj)
df
```

```python
#dataframe can be converted back to the json
print(df.to_json()) 
print(df.to_json(orient="records")) #it convert back to the json where the column become key and each row col value become value 
print(df.to_json(orient="index"))
```

```python
url ="https://microsoftedge.github.io/Demos/json-dummy-data/64KB.json"
df = pd.read_json(url) #as the .read_json accepts the file or url to the file so here we dont need the StringIO
df
```

```python
#reading the csv file from the url
url = "https://gist.githubusercontent.com/tijptjik/9408623/raw/b237fa5848349a14a14e5d4107dc7897c21951f5/wine.csv"
df = pd.read_csv(url)

df
```

```python
#we can convert back the dataframe into the csv file 
df.to_csv("wine.csv")
```
