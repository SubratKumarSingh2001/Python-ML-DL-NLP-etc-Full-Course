### Assignment 3: Data Cleaning



1. Create a Pandas DataFrame with 3 columns and 5 rows filled with random integers. Introduce some NaN values. Fill the NaN values with the mean of the respective columns.

```python
df = pd.DataFrame(np.random.randint(1,11, size=(5,3)), columns=['A','B','C'])
print("The Original Dataframe is: ")
print(df)

df.at[0,'A'] = np.nan
df.at[2,'B'] = np.nan
df.at[3,'C'] = np.nan
df.at[4,'B'] = np.nan

print("Dataframe after introducing the nan values: ")
print(df)

df.fillna(df.mean(), inplace=True) #It will fill the nan value with the mean of respective column
print("DataFrame with NaN values filled:")
df
```

### Assignment 4: Data Aggregation

1. Create a Pandas DataFrame with 2 columns: 'Category' and 'Value'. Fill the 'Category' column with random categories ('A', 'B', 'C') and the 'Value' column with random integers. Group the DataFrame by 'Category' and compute the sum and mean of 'Value' for each category.

```python
df = pd.DataFrame({'Category': np.random.choice(['A','B','C'], size=12), 'Value': np.random.randint(1,11, size=12)})
print("The Original Dataframe is")
print(df)

grouped_df = df.groupby('Category')['Value'].agg(['sum', 'mean'])
print("The mean and sum of each category value is: ")
print(grouped_df)
```

### Assignment 6: Time Series Analysis

1. Create a Pandas DataFrame with a datetime index and one column filled with random integers. Resample the DataFrame to compute the monthly mean of the values.

```python
date_range = pd.date_range(start="2025-01-01", end="2025-12-31", freq='D') #here D means Daily means every date from start and end will be considered
df = pd.DataFrame(date_range, columns=['Date'])
df['Values'] = np.random.randint(1,21, size=len(date_range))
df.set_index('Date', inplace=True)
print(df)

monthly_mean = df.resample('ME').mean()
print("Computing the mean of values for each month")
print(monthly_mean)
```

2. Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31' and one column filled with random integers. Compute the rolling mean with a window of 7 days.

```python
date_range = pd.date_range(start="2021-01-01", end="2021-12-31", freq='D') #here D means Daily means every date from start and end will be consider
df = pd.DataFrame(date_range, columns=['Date'])
df['Values'] = np.random.randint(1,21, size=len(date_range))
df.set_index('Date', inplace=True)
print(df)

rolling_mean = df.rolling(window=7).mean()
print("Computing the mean of values for every week(window of 7days) from start and end date")
print(rolling_mean)
```

2. Create a Pandas DataFrame with MultiIndex consisting of 'Category' and 'SubCategory'. Fill the DataFrame with random data and compute the sum of values for each 'Category' and 'SubCategory'.

```python
array = [['A','A','B','B'], ['one','two','one','two']]
index = pd.MultiIndex.from_arrays(array, names=('Category','Sub-Category'))
df = pd.DataFrame(np.random.randint(1,21, size=(4,3)), index=index, columns=['Value1','Value2','Value3'])
print(df)

print("The sum of values(value1, value2 and value3) for each category: ")
sum_values1 = df.groupby('Category').sum()
print(sum_values1)

print("The sum of values(value1, value2 and value3) for each subcategory: ")
sum_values2 = df.groupby('Sub-Category').sum()
print(sum_values2)
```

### Assignment 8: Pivot Tables

1. Create a Pandas DataFrame with columns 'Date', 'Category', and 'Value'. Create a pivot table to compute the sum of 'Value' for each 'Category' by 'Date'.

```python
date_range = pd.date_range(start="2021-01-01", end="2021-12-31", freq='D')
df = pd.DataFrame({'Date': np.random.choice(date_range, size=10), 'Category': np.random.choice(['A','B','C'], size=(10)), 'Value': np.random.randint(1,21, size=(10))})
print(df)

#Pivot Table: We perform data grouping and data aggregation, the result generated by grouping and aggregation is represented in tabular format i.e Pivot Table
pivot_table = pd.pivot_table(values='Value', index='Date', columns='Category', aggfunc='sum', data=df)
pivot_table
```

2. Create a Pandas DataFrame with columns 'Year', 'Quarter', and 'Revenue'. Create a pivot table to compute the mean 'Revenue' for each 'Quarter' by 'Year'.

```python
df = pd.DataFrame({'Year': np.random.choice(['2021','2022','2023','2024'], size=15), 'Quarter': np.random.choice(['Q1','Q2','Q3','Q4'], size=15), 'Revenue': np.random.randint(1,1000, size=15)})
print(df)

pivot_table = pd.pivot_table(values='Revenue', index='Year', columns='Quarter', aggfunc='mean', data=df)
pivot_table
```

### Assignment 10: Working with Text Data

1. Create a Pandas Series with 5 random text strings. Convert all the strings to uppercase.

```python
series = pd.Series(['subrat','raj','john','sagar','shiv'], index=['A','B','C','D','E'])
print("Original Series: ")
print(series)

new_series = series.str.upper() #.str consider the string and .upper() convert each index string value to uppercase and new_series is another series of string with uppercase values
print("New Series is: ")
print(new_series)
```

2. Create a Pandas Series with 5 random text strings. Extract the first three characters of each string.

```python
series = pd.Series(['subrat','raj','john','sagar','shiv'], index=['A','B','C','D','E'])
print("Original Series: ")
print(series)

first_three_char = series.str[:3] #.str consider the string and [:3] tell to extract first 3 char of each index string i.e char at 0,1 and 2 indexes. first_three_char is another series of string with first 3 chars
print(first_three_char, type(first_three_char))
```
