### Numpy

It is widely used Python Library which is used for the Scientific Computation in Python. It is used to create the arrays and matrices, along with some collection of mathematical functions applies on this data structures.

In this we will study about the Numpy its arrays and matrices and vectorized operations

```python
import numpy as np

#creating a 1-D array
arr = np.array([1,2,3,4])
print(arr, type(arr)) #type will tell the .ndarray which tells it is n-dimensional array

#shape of the array: tell the dimension of array
print(arr.shape)

#convert 1-D to 2-D array
arr1 = arr.reshape((1,4))
print(arr1, type(arr1))
```

```python
#creating a 2D array
arr2 = np.array([[1,2,3,4,5,6], [2,3,4,5,6,2]])
print(arr2, type(arr2))

#check the shape of arr2 
print(arr2.shape)
```

```python
#If we want to create a numpy array, same like we do list(1,10,2), tuple(1,10,2)
arr = np.arange(0,10).reshape(10,1) #10 rows and 1 columns reshaping
print(arr)
```

```python
#Some functions that are present in numpy and are useful
np.ones((2,3)) #creating an array of 2 rows and 3 cols and all elements are 1
```

```python
#identity matrix
np.eye(4) #It will create 4x4 identity matrix
```

```python
#Attributes of Numpy arrays

arr = np.array([[1,2,3], [4,5,6]])

print("Array is: \n", arr)
print("The Shape of array: ", arr.shape)
print("Dimension: ", arr.ndim)
print("Size(No.of Elements): ", arr.size)
print("Data type: ", arr.dtype)
print("Item size (in bytes): ", arr.itemsize)
```

```python
#Numpy vectorized operation 
arr1 = np.array([1,2,3,4,5,6])
arr2 = np.array([4,6,7,8,8,5])

#Element Wise Addition
print("Addition: ", arr1 + arr2)

#Element Wise Subtraction 
print("Subtraction: ", arr1 - arr2)


#Element Wise Subtraction 
print("Multiplication: ", arr1 * arr2)


#Element Wise Division
print("Division: ", arr1/arr2)
```

```python
# Universal Functions: Functions that are applied to every element of the array
arr1 = np.array([1,2,3,4])

print(np.sqrt(arr1))
print(np.exp(arr1))
print(np.sin(arr1))
print(np.log(arr1))
```

```python
#Performing Array slicing and Indexing
arr1 = np.array([[1,2,3,4,5], [1,3,4,5,7], [1,4,5,6,6]])

print("Array: \n", arr1)
print(arr1.shape)
```

```python
arr1[0][2] #accessing the 1st row and 3 col element
```

```python
arr1[1:,3:]
```

```python
arr1[:2,1:4]
```

```python
arr1[1:,2:4]
```

```python
#Modify array elements
arr1[0][0] = 100
print(arr1)

arr1[1:] = 100
print(arr1)
```

```python
#Statistical Concepts- Normalization
data = np.array([1,2,3,4])

#to have the mean of 0 and standard deviation of 1
mean = np.mean(data)
std_dev = np.std(data) 

#Normalize the data 
normalized_data = (data - mean) / std_dev
print("Normalized Data: ", normalized_data, type(normalized_data))
```

```python
#Performing Statistical Operation in Numpy
data = np.array([1,2,3,4])

mean = np.mean(data)
print("Mean: ", mean)

std_dev = np.std(data)
print("Standard Deviation: ", std_dev)

median = np.median(data)
print("Median: ", median)

variance = np.var(data)
print("Variance: ", variance)
```

```python
#Logical Operation
arr1 = np.array([1,2,4,5,5,6,8,9])

arr1[(arr1>4) & (arr1<8)]
```
