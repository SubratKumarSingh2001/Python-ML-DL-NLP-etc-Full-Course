# Data Visualization 

Matplotlib is the powerful plotting library for Python that enables the creation of static, animated and interactive visualizations. It is widely used for the Data Visualization. It is widely used for data visualization in data science and analytics. in this lesson, we will cover the basic of Mathplotlib, including creating various types of plots and customizing them

```python
import matplotlib.pyplot as plt
```

```python
x = [1,3,4,5,6,7,9]
y = [1,3,4,4,5,6,7]

#create the line plot
plt.plot(x,y) #It will create a memory location of an object and describe the type of plot
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic 2D Plot")
plt.show() #It will show rhe plot
```

```python
x = [1,3,4,5,6,7,9]
y = [1,3,4,4,5,6,7]

#create a customized line plot
plt.plot(x,y, color="red", linestyle="--", marker='o', linewidth=2, markersize=9)
plt.grid(True)
plt.show()
```

```python
#Suppose we want the multiple plots 
x = [1,2,3,4,5]
y = [5,6,7,8,9]
z = [3,5,6,7,8]

#now, .figure() will create a plotting area where all the plot will be shown i.e canvas, .figuresize=(9,6) 9 width and 6 height
plt.figure(figsize=(9,6))

#now we will create the subplot which will be plotted in this created canvas
plt.subplot(2,2,1) #means there are 2 rows and 2 columns and place this subplot at 1 cell
plt.plot(x,y, color="Red", marker="o")
plt.title("Plot-1")

plt.subplot(2,2,2) #means there are 2 row 2 col and place this plot at 2 cell
plt.plot(x,z, color="green", marker="o")
plt.title("Plot-2")

plt.subplot(2,2,3) #means there are 2 row 2 col and place this plot at 3 cell
plt.plot(y,z, color="blue", marker="o")
plt.title("Plot-3")

plt.subplot(2,2,4) #means there are 2 row 2 col and place this plot at 4 cell
plt.plot(z,x, color="purple", marker="o")
plt.title("Plot-4")
```

```python
#Let us understand the Bar Plot

categories = ['A', 'B', 'C', 'D', 'E']
values = [5,3,7,2,9]

#create a bar plot
plt.bar(categories, values, color='purple')
plt.show()
```

```python
#Histogram: are used to see the distribution of the dataset.They divide the data into bins(bars) and count the frequency(no.of data points) in each bins
data = [1,3,3,4,5,1,1,3,4,5,6,6,6,6,7,8,9,9,9]
plt.hist(data, bins=5, color="orange", edgecolor="black")
```

```python
#Creating a Scatter plot

x = [1,2,3,4,5]
y = [5,6,7,8,9]

plt.scatter(x, y, color="purple", marker="x")
```

```python
#Pie charts 
labels = ['A', 'B', 'C', 'D']
sizes = [20,50,10,30]
color = ['red', 'purple', 'orange', 'green']
explode = (0.1,0,0,0) #move out the first slice 

plt.pie(sizes, labels=labels, colors=color, autopct="%1.1f%%", shadow=True, explode=explode)
plt.show()
```

```python
#Lets do some practical: Sales Data Visualization
import pandas as pd
sales_df= pd.read_csv("data.csv")
sales_df.head(20)
```

```python
sales_df.info()
```

```python
sales_df = sales_df.rename(columns={'Make':"Car Company"})
sales_df.info()
```

```python
#Total odometer of each car company
total_odometer_company = sales_df.groupby('Car Company')['Odometer (KM)'].sum()
total_odometer_company
```

```python
total_odometer_company.plot(kind='pie', color='blue', autopct="%1.2f%%")
```

```python
total_odometer_company.plot(kind='bar', color='blue')
```

```python
new_total_odometer = total_odometer_company.reset_index()
new_total_odometer
```

```python
plt.plot(new_total_odometer['Car Company'], new_total_odometer['Odometer (KM)'], color="orange")
plt.xlabel('Car Company')
plt.ylabel('Odometer (KM)')
plt.show()
```
