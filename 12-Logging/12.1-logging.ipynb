# Logging

Loggers are the crucial for python applications, as loggers are the object that are used to keep track of what a program is doing i.e record messages - error, warning, normal executions, info about program etc,instead of using the print statements. Logging System also allow to save all the record messages i.e log messages within the .log file 

In this lesson, we will learn about the basics of logging, including how to configure logging, log levels and best practices for using logging in Python applications

```python
import logging 

#Configure(setting up the things to achieve the behavior of logging) the basic logging settings

'''
Levels tells python how minute or precisely you want to track you program. Means what minimum level of importance messages you want to track. If level=Debug means i want every small minute 
message related to info, errors, warning, critical) and if we choose level=Info means info(lowest) and critical (highest). means DEBUG message is hidden, and all else are considered. According
to the level decided, those all messages will be saved within the log file. If level=Error then error and critical messages will be stored within the log file
'''
logging.basicConfig(level=logging.DEBUG) 

# log messages with different severity levels
logging.debug("This is Debug message")
logging.info("This is info message")
logging.warning("This is warning messages")
logging.error("This is error message")
logging.critical("This is critical message")
```

# Log Levels

Python's logging module have several log levels indicating the severity of events. The default levels are - 
 
- DEBUG: Detailed information, typically of interest only when diagnosing problems.
- INFO: Confirmation that things are working as expected.
- WARNING: An indication that something unexpected happened or indicative of some problem in the near future. The software is still working as expected
- ERROR: Due to a more serious problem , the software has not been able to perform the way it is expected.
- CRITICAL: A very serious error, indicating that the program itself may be unable to continue running 

```python
#Configuring the logging
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force=True
)

# log messages with different severity levels
logging.debug("This is Debug message")
logging.info("This is info message")
logging.warning("This is warning messages")
logging.error("This is error message")
logging.critical("This is critical message")
```
