### Assignment 2: Logging with Different Handlers

1. Write a Python function to create a logger that logs messages to both a file named `app.log` and the console.

```python
def logger_with_handlers() :
    #creating the custom logger and setting its level
    logger = logging.getLogger('app')
    logger.setLevel(logging.DEBUG)
    logger.propagate = False

    if not logger.handlers:
        #create the handlers for custom logger i.e file and console handlers
        file_handler = logging.FileHandler('app.log')
        console_handler = logging.StreamHandler()

        #now we will set the specific format for custom logger and ensure that file and console logger follow that format
        formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
        file_handler.setFormatter(formatter)
        console_handler.setFormatter(formatter)

        #now we will set the level of each handler too
        file_handler.setLevel(logging.DEBUG)
        console_handler.setLevel(logging.DEBUG)

        #now we will connect the custom logger with the handlers we have created to ensure whenever the message is created using custom logger it should know that where the messages need to go
        logger.addHandler(file_handler)
        logger.addHandler(console_handler)
    
    logger.debug("This is Debug message")
    logger.info("This is info message")
    logger.warning("This is warning messages")
    logger.error("This is error message")
    logger.critical("This is critical message")

logger_with_handlers()
```

### Assignment 4: Rotating Log Files

1. Write a Python function to create a logger that uses a rotating file handler, which creates a new log file when the current log file reaches a certain size.

```python
def rotating_file_handler() :
    logger = logging.getLogger('rotating-handler-app')
    logger.setLevel(logging.DEBUG)

    if not logger.handlers :
        ''' 
        Creating the rotating handlers tells that if the file have reached the maximum file size make the current log file as old log file and create another log file with the same 
        name, and this will be current log fil and if this also become more than max size than then follow the above steps again and, also we can give how much backup file with the same 
        name we want to store if backup files count exceed the given number than the oldest file will be permanently deleted
        '''
        rotating_handler = logging.handlers.RotatingFileHandler('rotating-handler-app.log', maxBytes=2000, backupCount=5)
        formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s') #setting the format for this handler
        rotating_handler.setFormatter(formatter)

        logger.addHandler(rotating_handler)

    for i in range(100) :
        logger.debug(f"This is debug message {i}")

rotating_file_handler()
```

```python
def logging_exception() :
    logger = logging.getLogger('exception-logger')
    logger.setLevel(logging.ERROR)

    if not logger.handlers :
        file_handler = logging.FileHandler('exception-logger.log')
        formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
        file_handler.setFormatter(formatter)

        logger.addHandler(file_handler)

        try : 
            result = 1/0
            print(result)
        except ZeroDivisionError as e :
            logger.exception(f"There is an exception as {e}")

logging_exception()
```


### Assignment 6: Contextual Logging


1. Write a Python function to create a logger that includes contextual information (e.g., function name, line number) in the log messages.

```python
#Contextual Logging: means providing more information i.e metadata to logs, so that the debugging could become easier

def contextual_logger() :
    logger = logging.getLogger('contextual-logger')
    logger.setLevel(logging.DEBUG)

    if not logger.handlers :
        file_handler = logging.FileHandler('contextual-logger.log')
        # %(funcName)s: tells which function calls the log messages, %(lineno)d: tells the log message was written at what line of python code
        formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s - %(funcName)s - %(lineno)d')
        file_handler.setFormatter(formatter)

        logger.addHandler(file_handler)

    def test() :
        logger.debug("This is Debug message")
        logger.info("This is info message")
        logger.warning("This is warning messages")
        logger.error("This is error message")
        logger.critical("This is critical message")

    test()    

contextual_logger()
```

2. Modify the function to include additional contextual information (e.g., user ID, session ID).

```python
def addition_contextual_info(user_id, session_id) :
    logger = logging.getLogger('add-contextual-info')
    logger.setLevel(logging.DEBUG)

    if not logger.handlers :
        file_handler = logging.FileHandler('add-contextual-info.log')
        formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s - %(funcName)s - %(lineno)d - User-id: %(user_id)s - Session-id: %(session_id)s')
        file_handler.setFormatter(formatter)

        logger.addHandler(file_handler)

    extra = {'user_id': user_id, 'session_id': session_id}

    logger.debug("This is Debug message", extra=extra)
    logger.info("This is info message", extra=extra)
    logger.warning("This is warning messages", extra=extra)
    logger.error("This is error message", extra=extra)
    logger.critical("This is critical message", extra=extra)    


addition_contextual_info(32, 28)
```

### Assignment 7: Configuring Logging with a Dictionary

There are 3 different ways we can config our loggers 
1. BasicConfig: configuration for the root log
2. Manual Config: configuration for each custom log
3. DictConfig: creating a python dictionary to config the log

We have studied till now first and second way. Out of all three the best way is dictionary configuration because you describe everything in one Python dictionary whether it is logger, handlers 
and formatter at once instead of writing many Python lines to set up loggers, handlers, and formatters.

Without dictionary: 
- logger = logging.getLogger(...)
- handler = logging.FileHandler(...)
- formatter = logging.Formatter(...)
- handler.setFormatter(formatter)
- logger.addHandler(handler)

With dictionary:
- LOGGING_CONFIG = { ... }

How to use it:
- logging.config.dictConfig(LOGGING_CONFIG)

With this Python Configures
- Loggers
- Handlers
- Formatter 
automatically

Below is the practical implementation of dictConfig 



1. Write a Python function to configure logging using a dictionary. The configuration should include handlers for both file and console logging.

```python
import logging.config

def dict_config() : 
    log_config = {
        #version of config
        'version':1,
        #formatter for root log
        'formatters': {
            'detailed' : {
                'format': '%(asctime)s - %(name)s - %(levelname)s - %(message)s'
            },
            'simple' : {
                'format': '%(asctime)s - %(levelname)s - %(message)s'
            }
        },
        #handlers for the root log
        'handlers': {
            'file' : {
                'class':'logging.FileHandler',
                'filename':'dict-config.log',
                'formatter':'detailed', 
                'level':'DEBUG'
            },
            'console' : {
                'class':'logging.StreamHandler',
                'formatter':'simple',
                'level':'DEBUG'
            }
        },
        #Create the root logger
        'root' : {
            'handlers':['file','console'],
            'level':'DEBUG'
        }
    }

    logging.config.dictConfig(log_config)
    logger = logging.getLogger('')
    logger.debug('This is a debug message')
    logger.info('This is an info message')
    logger.warning('This is a warning message')
    logger.error('This is an error message')
    logger.critical('This is a critical message')


dict_config()
```

3. Create a dict log file which have custom log and root log along with there format, handlers 

```python
def custom_dict_config() :
    custom_dict_config = {
        "version": 1,

        # ---------- FORMATTERS ----------
        "formatters": {
            'detailed' : {
                'format': '%(asctime)s - %(name)s - %(levelname)s - %(message)s'
                },
            'simple' : {
                'format': '%(asctime)s - %(levelname)s - %(message)s'
            }
        },

        # ---------- HANDLERS ----------
        "handlers": {
            # Root logger handler
            "root_handlers": {
                "class": "logging.FileHandler",
                "filename": "root.log",
                "formatter": "simple",
                "level": "DEBUG"
            },

            # Custom logger handler
            "app_handlers": {
               "class": "logging.FileHandler",
                "filename": "app-file.log",
                "formatter": "detailed",
                "level": "DEBUG"
            },

            #Common console handler for both root and custom log
            "console": {
                "class": "logging.StreamHandler",
                "formatter": "detailed",
                "level": "DEBUG"
            }
        },

        # ---------- ROOT LOGGER ----------
        "root": {
            "handlers": ["root_handlers", "console"],
            "level": "DEBUG"
        },

        # ---------- CUSTOM LOGGERS WHICH ARE DEFINED INSIDE loggers ----------
        "loggers": {
            "app-file": {
                "handlers": ["app_handlers", "console"],
                "level": "DEBUG",
            }
        }
    }

    logging.config.dictConfig(custom_dict_config)

    # custom logger
    logger = logging.getLogger("app-file")
    logger.debug('This is a debug message')
    logger.info('This is an info message')

    #root logger
    logging.debug('This is a debug message')
    logging.info('This is an info message')


custom_dict_config()
```

### Assignment 9: Logging Performance


1. Write a Python script to benchmark the performance of logging with different handlers (e.g., file handler, console handler, rotating file handler).

```python
import time

def logging_performance() :
    logger = logging.getLogger('performance-logger')
    logger.setLevel(logging.DEBUG)

    #Using the FileHandler
    file_handler = logging.FileHandler('performance-logger.log')
    file_handler.setLevel(logging.DEBUG)
    logger.addHandler(file_handler)

    start_time = time.time()
    for i in range(100) :
        logger.debug("This is the debug message")
    end_time = time.time()
    print("The performance of File Handlers in Seconds is: {}".format(end_time - start_time))
    logger.removeHandler(file_handler)

    #Using the StreamHandler
    console_handler = logging.StreamHandler()
    console_handler.setLevel(logging.DEBUG)
    logger.addHandler(console_handler)

    start_time = time.time()
    for i in range(100) :
        logger.debug("This is the debug message")
    end_time = time.time()
    print("The performance of Stream Handlers in Seconds is: {}".format(end_time - start_time))
    logger.removeHandler(console_handler)

    #Using Rotating Handler
    rotating_handler = logging.handlers.RotatingFileHandler('performance-rotating.log', maxBytes=2000, backupCount=5)
    rotating_handler.setLevel(logging.DEBUG)
    logger.addHandler(rotating_handler)

    start_time = time.time()
    for i in range(100):
        logger.debug('This is a debug message')
    end_time = time.time()
    print('The performance of Rotating Handlers in Seconds is: {} seconds'.format(end_time - start_time))
    logger.removeHandler(rotating_handler)

logging_performance()
```


2. Modify the script to compare the performance of logging with and without message formatting.
