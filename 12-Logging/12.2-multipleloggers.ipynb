# Logging with Multiple Loggers

You can create multiple loggers for different parts of your application

```python
import logging

#.getloggers() is creating logging system for each module i.e .py file Get (or create) a logger object for a particular module, and use that object to send log messages from that module.

logger1 = logging.getLogger('app1')
logger1.setLevel(logging.DEBUG)

logger2 = logging.getLogger('app2')
logger2.setLevel(logging.WARNING)


logging.basicConfig(
    filename='multiple.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force=True
)
```

```python
logger1.debug("This is module1 debug message")
logger2.error("This is logger2 info message")
logger2.critical("This is logger2 info message")
```
