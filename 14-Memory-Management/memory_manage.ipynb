# Reference Counting 

Reference Counting is the primary method Python uses to manages the memory. Each object in Python maintains a count of the references pointing to it. When the reference count drops to zero, the memory occupied by the object is deallocated.


```python
# Reference Counting 

import sys

a = []
#here output is 2 because one reference is coming from 'a' variable and other from the getrefcount() method
print(sys.getrefcount(a))
```

```python
b = a 
#here output is 3 because one reference is coming from 'a' variable, second from the getrefcount() method and third from variable 'b'
print(sys.getrefcount(b))
```

```python
del b 
#output 2 because 2 is deleted 
print(sys.getrefcount(a))
```

### Garbage Collection

Python includes a cyclic garbage collector to handle reference cycles. Reference cycles occur when objects reference each other, preventing their references count from reaching zero


```python
import gc 

# enable the garbage collection 
gc.enable()
```

```python
gc.disable() #to disable the garbage collection 
```

```python
#Here we manually trigger the garbage collector, it helps in finding the unused objects find and free the space and basically remove the reference cycles means a refer to b and vice versa
gc.collect()
```

```python
# want to get the garbage collection statistics
print(gc.get_stats())
```

```python
# get the unreachable objects 
print(gc.garbage)
```

### Efficient Memory Management Best Practices

1. Use Local Variables: Local Variables have a shorter lifespan and are freed sooner than global variables.
2. Avoid Circular Referencing: Circular Reference can lead to memory leaks if not properly managed.
3. Use generators: Generators ensure to yield the one output at a time by keeping one item in memory at a time, making it memory efficient.
4. Explicitly Delete the objects: use the del statement to delete variables and objects explicitly.
5. Profile Memory Usage: Use memory profiling tools like tracemalloc and memory_profiler to identify memory leaks and optimize memory usage

```python
#Handing Circular Reference

import gc

class MyObject : 
    def __init__(self, name) :
        self.name = name
        print(f"Object {self.name} is created")

    def __del__(self) :
        print(f"Object {self.name} deleted")

# create a circular reference 
obj1 = MyObject("obj1")
obj2 = MyObject("obj2")

obj1.ref = obj2
obj2.ref = obj1

#Till now this will not be deleted because due to circular reference means __del__ dunder method will not get executed 
del obj1
del obj2

# we need to manually trigger the garbage collector to find the unused objects and circular reference and remove those
gc.collect()
```

```python
#Generators for memory efficient 
def generate_num(n) :
    for i in range(n) :
        yield i

for i in generate_num(40000) :
    print(i)
    if i > 10 :
        break   
```

```python
#Profiling Memory Usage with tracemalloc : used to track the memory usage 
import tracemalloc

def create_list() :
    return [i for i in range(1000)]

def main() :
    tracemalloc.start()

    create_list()
    
    snap_shot = tracemalloc.take_snapshot()
    top_stats = snap_shot.statistics('lineno')

    for stats in top_stats[::] :
        print(stats)
```

```python
main()
```
