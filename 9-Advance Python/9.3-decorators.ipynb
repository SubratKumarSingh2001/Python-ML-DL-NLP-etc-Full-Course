### Decorators

Decorators are the advance python concept, which provide the flexibility of changing the behavior of functions or the class methods. In simple we can say that adding functionalities to the function without changing the actual code. Means we call a function and in return we get another function.

We can also say that decorators are like passing a gift to a function and asking to decorate this gift i.e wrap it in a box and decorate it well and then return this gift.

Syntax of Decorator: @decorator_name

To understand the decorator we need to first understand the 1.Function copy 2.Closure and 3.Decorators

```python
#1. Function Copy

#Suppose we have created a function 
def message(guest="Guest") :
    print(f"Welcome! {guest} to the course")

#calling the function
print(message("Subrat"))
print(message) #It will be showing the memory location of the function

#now we want to create the copy of the function
mess = message #here we are assigning the function to a variable and with the help of variable name i.e mess() we can also call the same function
print(mess())
print(mess) #here we can see that message and mess showing the same memory location of the function 
```

```python
print(mess())
#suppose, we delete the message function 
del message
#now check whether we are still able to print the mess() function
print(mess())
```

```python
#2. Function Closure: Creating a function inside the function i.e nested functions

def message(func) :
    mess = "Welcome" #local variable
    def sub_message() :
        print("Welcome to the advance python topics")
        print(mess) #mess is a local variable created inside the main function i.e message(), so it can be accessed anywhere inside the function
        print(func("Welcome to the course lets dive deeper in it.")) #if the func is len then it is like len(Whatever the message inside it)
        print("Please learn these topics")
    return sub_message
```

```python
# now what if we pass function as an input parameter
x = message(print) #here the print is an in-built function that we have passed 
x() #means here x and sub_message point to same function i.e sub_message() means if we call x() we will be indirectly calling sub_message()
```

```python
#3. Decorators

def message(func) :
    mess = "Welcome" #local variable
    def sub_message() :
        print("Welcome to the advance python topics")
        func()
        print("Please learn these topics")
    return sub_message

'''
Suppose we are calling this method and want to add some additional functionalities and behavior to it. So the message() func use this greet() and wrap 
it with another function sub_message() which is declared inside the message().
In short by calling greet() method indirectly we call sub_message() with the help of the message()
'''
def greet() :
    print("Hello, Welcome to the course")

#This we are simply calling the greet method
# greet()
# print() 
#Here we are passing a function as an argument, which defines the decorators
greet = message(greet)
# print(greet) #greet and sub_message have same memory address of the sub_message() function
greet()
```

```python
def message(func) :
    mess = "Welcome" #local variable
    def sub_message() :
        print("Welcome to the advance python topics")
        func()
        print("Please learn these topics")
    return sub_message()


#if we want to call message function and pass greet function as an argument without using a way message(greet), we can use @method_name
'''
@message 
def greet :
    print("Hello, Welcome to the course") 
The above is same as greet = message(greet) where we are passing the greet function as an input argument to the decorator message. Here we have used
return sub_message() directly so once the decorator is called sub_message() calls and returns its result. '''

@message
def greet() :
    print("Hello, Welcome to the course")
```

```python
def message(func) :
    mess = "Welcome" #local variable
    def sub_message() :
        print("Welcome to the advance python topics")
        func()
        print("Please learn these topics")
    return sub_message

@message
def greet() :
    print("Hello, Welcome to the course")

#suppose we want to call the sub_message() later on when we call the greet() function 
'''Here we have called the greet() function which is same as above now when we call the greet() function indirectly we are calling the sub_message() 
function, where the decorator take this greet() function and add some functionalities to it. 

greet = message(greet) when this message(greet) called it return sub_message object of sub_message() function. Now when we call greet() as 
greet = sub_message then greet() changes to sub_message() this is the reason we have return the sub_message instead of sub_message(). Now we can 
use greet() function anywhere according to the requirements.'''
greet()
```

```python
#Decorators with arguments

def repeat(n) :
    def decorator(func) :
        def wrapper(*args, **kwargs) :
            for i in range(n) :
                result = func(*args, **kwargs)
                print(result)
        return wrapper
    return decorator

'''Here the repeat(5) is the normal function which is executed first i.e sum = repeat(5)(sum) and repeat(5) return the decorator repeat(5) was used to
have an argument n which will be used inside the function.

Now the repeat(5) return the decorator it become sum = decorator(sum) now it is decorator which take function as an argument and returns another function
wrapper. 

In short sum(2,5) we are calling the wrapper(2,5) function where 2,5 are the positional arguments. Now sum = decorator(sum) return the wrapper function
object named wrapper i.e sum = wrapper become same. 

So if we call sum(2,5) being sum = wrapper, it seems we are calling wrapper(2,5)
'''
@repeat(5) 
def sum(a, b) :
    return a + b

sum(2,3)                                                                                                                                                                                                                
```
