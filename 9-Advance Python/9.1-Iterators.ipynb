### Iterators 

Iterators are the advance python concept which we used to perform efficient looping and memory management. It is used to access the items of sequential iterators without showing the overall structure.

Suppose we have an a list and want to iterate through each element of the list so we will use the loop either for or while loop which shows are showing the overall structure of looping that how we are accessing the element of the list. Therefore we use the iterators

To use iterator we use iter() method and if we want to access the element of the iterator we will use next()

```python
#Let see some examples
my_list = [1,2,3,4,5,6]

#use of the iterator 
iterator = iter(my_list)
print(iterator)
```

```python
#now to access the elements from the iterator without exposing the overall structure that how it is happening 
next(iterator) #it follows lazy process and once all the elements are shown at last when empty it tells that hey iterator has no element StopIteration
```

```python
#to handle this error we can use the try and except block
try :
    print(next(iterator))
except StopIteration :
    print('Hey Stop no element is present inside the iterator') #This exception will be shown when there is no element present inside the iterable
```

```python
#We can perform this on the strings also
name = 'Subrat'
iterator_str = iter(name)
print(iterator_str)
try :
    print(next(iterator_str))
    print(next(iterator_str))
    print(next(iterator_str))
    print(next(iterator_str))
    print(next(iterator_str))
    print(next(iterator_str))
    print(next(iterator_str))
    #after this there is no element in the iterator return the exception
except StopIteration :
    print("name dont have any character left.. So stop the iteration")
```

```python
#Let us understand the iterator internal implementation

# my_list = [1,2,3,4,5]
# #now if, I want to iterate to list, we can use for or while loop
# for item in my_list : #this way we are exposing the overall structure 
#     print(item, end=" ")

'''
Behind the scene what Python does : 
1. Python call iter(my_list) to access each element of the list, internally it behaves my_list.__iter__() where my_list is instance of class list
2. Now __iter__() which is instance method will be called that return Iterator object i.e list_iterator telling the type of iterator. Now 
iterator = iter(my_list), here iterator holds to same object.
3. If we do next(iterator) which means same as iterator.__next__() calling next() method of iterator object which will return the first element 
of the list and internally it maintains a counter to keep track of which element is being accessed.
4. When all the elements are accessed and we again call next() it will raise StopIteration exception.
'''

#Let us verify the above steps
my_list = [10,20,30,40,50]

#Step 1
iterator = iter(my_list) #or iterator = my_list.__iter__()
print(iterator) #it will print the iterator object

#Step 2
print(type(iterator)) #it will print the type of iterator object

#Step 3
print(next(iterator)) #or print(iterator.__next__())
print(next(iterator))
print(next(iterator))   
print(next(iterator))
print(next(iterator))

#Step 4
try :
    print(next(iterator))
except StopIteration :
    print("No element is present inside the iterator.. So Stop the iteration")
```
