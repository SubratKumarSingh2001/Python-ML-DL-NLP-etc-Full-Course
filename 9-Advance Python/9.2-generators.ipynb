### Generators 

It is a special way to create iterators. Suppose you want to return the list of squares of numbers with the help of the function so in this case the whole extra list will be created and stored inside the memory and return it.

But the generators says that, I wont create an extra memory for the list instead produce one value at a time and return it which make it memory efficient.

This is achieved with the help of the yield keyword instead of using the return keyword it produce a value and along with that return that value.
yield keyword run the function generate the value and return it and pause the function and when required that function resumes from there. They are lazy produce the values when needed.

They are preferly used when dealing with files having large data instead of loading whole in memory at once it produce one line at a time


```python
#Suppose an example: you want to return the list of square of numbers 
```

```python
def square(n) :
    for i in range(n) :
        yield i**2

print(square(3)) #this tell it is a generator object because the moment python sees yield inside a function it become a generator function
```

```python
for i in square(3) : #now the function square(3) yield one value at a time, it produce series of values very lazily when needed
    print(i)
```

```python
a = square(3)
print(a)
```

```python
try : 
    print(next(a)) #same as of the iterator we can use next() to access the value
except StopIteration :
    print("Stop the iteration as no element is present")
```

```python
def my_generator() :
    yield 1 #first 1 will be yielded
    yield 2 #then 2 will be yielded
    yield 3 #then 3 will be yielded
 

generator = my_generator()
print(generator)
print(next(generator))
print(next(generator))
print(next(generator))
```

```python
#Real life practical example: When we want to read a large data file, instead loading the whole file in the memory we can produce one line at a time 
def read_file(filepath) :
    with open(filepath, 'r') as file :
        for line in file :
            yield line # this instead of loading whole file in the memory this will produce a line at a time 

#1st way to access each line of the file 
# read_file = read_file("large_file.txt")
# #This will yield one line at a time
# print(next(read_file))
# print(next(read_file))
# print(next(read_file))
# print(next(read_file))


#2nd way to access each line of the file 
filepath = "large_file.txt"
for line in read_file(filepath) :
    print(line.strip())
```

```python
def countDown(n) :
    current = n
    while True :
        current -= 1
        if current < 0 :
            return
        yield current # which will generate one value at a time 

countdown = countDown(5)
print(countdown)
# while True :
#     try :
#         print(next(countdown))
#     except StopIteration :
#         print("Stop the Iteration as no element is left to print")
#         break # because the except block handle the exception and because of while True, next(countdown) again runs and the StopIteration executed infinite times 

#using the for loop
for i in countdown : #for loop handle, that when the StopIteration is raised just stop the execution 
    print(i)
```
